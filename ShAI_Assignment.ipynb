# Machine Learning 

divide data frame to data and result 

```python
x = df[['BasePay', 'OvertimePay']].loc[:30000]
y = df[['TotalPay']].loc[:30000]
```

divide x and y to train and test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

creates a linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(xtrain,ytrain)
```

predict new values from test data

```python
ypred = regr.predict(xtest)
ypred
```

 coef_:the coefficients (slope) of the linear regression model for each feature.
 intercept_:the intercept (bias) of the linear regression model.

```python
print(regr.coef_)
print(regr.intercept_)
```

evaluate the model's performance 
caluclate accuracy using r2_score

```python
from sklearn.metrics import r2_score
acc=r2_score(ytest, ypred)
print(" Regression model accuracy (in %):", acc*100)
```

This summary will include various statistics, including coefficients, standard errors, t-values, p-values, and R-squared, providing insights into the relationships between the independent and dependent variables in your linear regression model.

```python
import statsmodels.api as sm

# Create an OLS model
X = sm.add_constant(xtrain)  # Add a constant term
model = sm.OLS(ytrain, xtrain)

# Fit the model to your data
results = model.fit()

# Get the summary
print(results.summary())
```

evaluate how well the model's predictions match the actual values in terms of absolute differences.

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(ytest, ypred))
```

evaluate how well the model's predictions match the actual values in terms of squared differences.

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(ytest,ypred))
```

It measures the average absolute percentage difference between the actual and predicted values

```python
from sklearn.metrics import mean_absolute_percentage_error
print(mean_absolute_percentage_error(ytest,ypred))
```

 It measures the average squared difference between the actual and predicted values, providing insight into the overall accuracy of model.

```python
print(mean_squared_error(y_true=ytest,y_pred=ypred,squared=False))
```
